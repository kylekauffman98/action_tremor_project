# Overview
Train classifier on x acceleration features only. Binary classification. Does not account for balancing data. Strictly reads in data and trains MLP classifier using all tsfel features

```python
import os
import tsfel
import numpy as np
import pandas as pd
```

### Set directory locations

```python
norm_dir = './Data/Normal_data'
tremor_4_dir = './Data/Tremor_added/4Hz'
tremor_8_dir = './Data/Tremor_added/8Hz'
```

### Dataframes for all 3 levels. 
Read in all files

```python
df_norm = pd.concat((pd.read_csv(norm_dir+'/'+f, header=0, names=["x", "y", "z"]) for f in os.listdir(norm_dir)), ignore_index=True)
```

```python
df_4hz = pd.concat((pd.read_csv(tremor_4_dir+'/'+f, header=0, names=["x", "y", "z"]) for f in os.listdir(tremor_4_dir)), ignore_index=True)
```

```python
df_8hz = pd.concat((pd.read_csv(tremor_8_dir+'/'+f, header=0, names=["x", "y", "z"]) for f in os.listdir(tremor_8_dir)), ignore_index=True)
```

```python
X_train_sig_norm = df_norm["x"].to_frame()
```

### Extract features for all 3 dataframes. Add labels after feature extraction. 0 for no tremor, 1 for tremor

```python
%%time
cfg_file_norm = tsfel.get_features_by_domain()                                                        # If no argument is passed retrieves all available features
X_train_norm = tsfel.time_series_features_extractor(cfg_file_norm, X_train_sig_norm, fs=20, window_size=200)
```

```python
X_train_norm['label'] = 0
```

```python
X_train_sig_4hz = df_4hz["x"].to_frame()
```

```python
%%time
cfg_file_4hz = tsfel.get_features_by_domain()                                                        # If no argument is passed retrieves all available features
X_train_4hz = tsfel.time_series_features_extractor(cfg_file_4hz, X_train_sig_4hz, fs=20, window_size=200)
```

```python
X_train_4hz['label'] = 1
```

```python
%%time
X_train_sig_8hz = df_8hz["x"].to_frame()
cfg_file_8hz = tsfel.get_features_by_domain()                                                        # If no argument is passed retrieves all available features
X_train_8hz = tsfel.time_series_features_extractor(cfg_file_8hz, X_train_sig_8hz, fs=20, window_size=200)
```

```python
X_train_8hz['label'] = 1
```

### Combine all 3 feature dataframes into one.

```python
df_x_features = pd.concat([X_train_norm, X_train_4hz, X_train_8hz], ignore_index=True)
```

```python
df_x_features
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = df_x_features.drop('label', axis=1).values
columns = df_x_features.drop('label', axis=1).columns
sc = StandardScaler()
```

### Use sklearn standard scaler. Apparently better than normalizing for classification.
Drops label and then adds back to avoid scaling the label.
Note: Normalization is better for regression tasks.

```python
%%time
x_scaled = sc.fit_transform(x)
df_x_features_scaled = pd.DataFrame(x_scaled, columns=columns)
df_x_features_scaled['label'] = df_x_features['label']
```

```python
df_x_features_scaled
```

```python
from sklearn.model_selection import train_test_split
```

### Divide into training and validation sets 70/30 split

```python
training_set_x, validation_set_x = train_test_split(df_x_features_scaled, test_size = 0.3, random_state = 21)
```

```python
X_train_x = training_set_x.iloc[:,0:-1].values
Y_train_x = training_set_x.iloc[:,-1].values
X_val_x = validation_set_x.iloc[:,0:-1].values
y_val_x = validation_set_x.iloc[:,-1].values
```

```python
def accuracy(confusion_matrix):
   diagonal_sum = confusion_matrix.trace()
   sum_of_all_elements = confusion_matrix.sum()
   return diagonal_sum / sum_of_all_elements
```

```python
from sklearn.neural_network import MLPClassifier
```

### Train classifier

```python
classifier = MLPClassifier(hidden_layer_sizes=(150,100,50), max_iter=300,activation = 'relu',solver='adam',random_state=1)
```

```python
%%time
classifier.fit(X_train_x, Y_train_x)
```

### Evaluate prediction

```python
y_pred_x = classifier.predict(X_val_x)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_pred_x, y_val_x)
```

```python
print("Accuracy of MLPClassifier : ", accuracy(cm))
```
